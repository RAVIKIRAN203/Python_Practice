```python
df1.groupby('Region').agg({'Sales':'sum','Profit':'mean'}).rename(columns={'Sales':'Total Sales','Profit':'Average Profit'}).reset_index()

# agrregations using agg function
```

```python
df1.groupby('Region').agg({'Sales':'sum','Sales':'mean'}).reset_index()
    
# in dictionary key should be unique , so it will take the last key value so mean is overriden by sum
```

```python
df1.groupby('Region').agg({'Sales':['sum','mean']}).reset_index()
# dictionary with multiple values for a key as list
```

```python
df1.groupby('Region').agg({'Sales':('sum','mean')}).reset_index()
# dictionary with multiple values for a key as tuple
```
